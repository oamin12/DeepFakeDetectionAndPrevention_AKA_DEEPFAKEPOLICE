```python
#loop over the images in the current directory and predict on them
# resize all images to 384*384 before predicting
# the model uses a path only, so we need to read the image, resize it, and then pass it to the model


import os
import cv2
from PIL import Image
import numpy as np

directory = "Test"
for filename in os.listdir(directory):
    if (filename.endswith(".jpg") or filename.endswith(".png")) and filename != "temp.jpg":
        img = cv2.imread(os.path.join(directory, filename))
        img = cv2.resize(img, (224, 224))
        # Convert BGR to RGB and pass directly as PIL Image
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        pil_image = Image.fromarray(img_rgb)

        result = DeepfakeDetViT(pil_image)

        print(f"{filename}: {result}")  
```
